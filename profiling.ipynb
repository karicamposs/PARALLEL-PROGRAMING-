```python
import timeit
import multipr

# Define la cantidad de veces que deseas ejecutar el código para calcular el tiempo promedio.
number_of_executions = 30

# Define la cantidad de subdivisiones para la aproximación de π.
n = 100000

# Define el número de procesos a utilizar.
num_processes = 1

# Define la función que ejecutará timeit
def run_code():
    multipr.approximate_pi(n, num_processes)

# Mide el tiempo de ejecución
execution_timem = timeit.timeit(run_code, number=number_of_executions)

# Imprime el tiempo de ejecución y el resultado de la aproximación de π
print(f"Execution time: {execution_timem} seconds")
print(f"Approximation of π: {multipr.approximate_pi(n, num_processes)}")
```

```python
import timeit
import dpce

number_of_executions = 30
n = 100000
num_processes = 1

def run_code():
    return dpce.approximate_pi(n, num_processes)

execution_timed = timeit.timeit(run_code, number=number_of_executions)
result = run_code()  # Ejecutar nuevamente para obtener el resultado de la aproximación

print(f"Execution time: {execution_timed} seconds")
if result is not None:
    print(f"Approximation of π: {result}")
```

```python
import matplotlib.pyplot as plt


categories = ['Non-Paralellization', 'Multiprocesing', 'Parallel dtributed']

# Valores promedio
averages = [execution_timen, execution_timem, execution_timed]

# Creación del gráfico de barras
fig, ax = plt.subplots()
ax.bar(categories, averages, color=['blue', 'green', 'red'])

# Añadir título y etiquetas
ax.set_title('Run time comparison')
ax.set_ylabel('Average time (seconds)')
ax.set_xlabel('Execution method')

# Mostrar valores en las barras
for i, v in enumerate(averages):
    ax.text(i, v + 0.05, f"{v:.2f} s", ha='center', color='black')

# Mostrar el gráfico
plt.show()
```
